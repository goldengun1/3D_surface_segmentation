```python
import numpy as np
from typing import List
import surface_functions as sf
```

```python
x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)
X, Y = np.meshgrid(x, y)
Zs = []
names = []

Zs.append(sf.surface1(X, Y))
names.append("Surface1 Function")
Zs.append(sf.surface2(X, Y))
names.append("Surface2 Function")
Zs.append(sf.himmelblau_function(X, Y))
names.append("Himmelblau Function")
Zs.append(sf.ackley_function(X, Y))
names.append("Ackley Function")
Zs.append(sf.rosenbrock_function(X, Y))
names.append("Rosenbrock Function")
Zs.append(sf.rastrigin_function(X, Y))
names.append("Rastrigin Function")
Zs.append(sf.michalewicz_function(X, Y))
names.append("Michalewicz Function")
Zs.append(sf.goldstein_price_function(X,Y))
names.append("Goldstein Price Function")
```

```python
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(24,30))
for i, (Z, name) in enumerate(zip(Zs, names)):
    ax = fig.add_subplot(int(len(Zs) / 2),3, i+1, projection='3d')
    ax.set_title(label=name)
    ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

plt.show()
```
